## Problem Statement

Happy Customer Bank is a mid-sized private bank that deals in all kinds of banking products, like Savings accounts, Current accounts, investment products, credit products, among other offerings.The bank also cross-sells products to its existing customers and to do so they use different kinds of communication like tele-calling, e-mails, recommendations on net banking, mobile banking, etc. In this case, the Happy Customer Bank wants to cross sell its credit cards to its existing customers. The bank has identified a set of customers that are eligible for taking these credit cards.Now, the bank is looking for your help in identifying customers that could show higher intent towards a recommended credit card,they have given dataset in `train.csv` given:

> Customer details (gender, age, region etc.)

> Details of his/her relationship with the bank (Channel_Code,Vintage, 'Avg_Asset_Value etc.)

### Handling Outliers and Skewness of Avg Account Balance

```python
df_train['Avg_Account_Balance'] = np.log(df_train['Avg_Account_Balance'])
sns.kdeplot(df_train['Avg_Account_Balance'])
plt.show()
```

```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
sns.boxplot(x=df_train['Avg_Account_Balance'])
```

```python
low_avg_sal , upper_avg_sal = remove_outliers(df_train["Avg_Account_Balance"])
df_train["Avg_Account_Balance"] = np.where(df_train["Avg_Account_Balance"]>upper_avg_sal,upper_avg_sal,df_train["Avg_Account_Balance"])
df_train["Avg_Account_Balance"] = np.where(df_train["Avg_Account_Balance"]<low_avg_sal,low_avg_sal,df_train["Avg_Account_Balance"])
```

```python
predictor_test = pd.concat([Test_categorical,pd.DataFrame(Test_numerical)],axis=1)
```

## Modelling

```python
predictor_train = train_new_df.drop(['Is_Lead'],axis=1)
target_train    = train_new_df['Is_Lead']
```

```python
submission_df = pd.read_csv("sample_submission.csv")
print("Submission file is loaded")
```

```python
#Creation ML Model,Fine Tune with MODEL Evaluation METRIC - "ROC_AUC" with k-FOLD Cross Validation and Predict Target "Response"
kf=KFold(n_splits=10,shuffle=True)

preds   = list()
y_pred  = []
rocauc_score = []

for i,(train_idx,val_idx) in enumerate(kf.split(predictor_train)):    
    
    X_train, y_train = predictor_train.iloc[train_idx,:], target_train.iloc[train_idx]    
    X_val, y_val = predictor_train.iloc[val_idx, :], target_train.iloc[val_idx]
   
    print('\nFold: {}\n'.format(i+1))

    lg = LGBMClassifier(metric = 'auc', 
                       n_estimators=50000,    
                       bagging_fraction=0.95, 
                       subsample_freq = 2, 
                       objective ="binary",
                       importance_type = "gain",
                       verbosity = -1,
                       random_state=294,
                       num_leaves = 300,
                       boosting_type = 'gbdt',
                       learning_rate=0.15,
                       max_depth=4, 
                       scale_pos_weight=2,
                       n_jobs=-1 
                      )

    lg.fit(X_train, y_train, eval_set=[(X_train, y_train),(X_val, y_val)] ,early_stopping_rounds=100 ,verbose=100)

    roc_auc = roc_auc_score(y_val,lg.predict_proba(X_val)[:, 1])
    rocauc_score.append(roc_auc)
    preds.append(lg.predict_proba(predictor_test[predictor_test.columns])[:, 1])
    
y_pred_final = np.mean(preds,axis=0)    
submission_df['Is_Lead']=y_pred_final
```

```python
print('ROC_AUC - CV Score: {}'.format((sum(rocauc_score)/10)),'\n')
print("Score : ",rocauc_score)
```
